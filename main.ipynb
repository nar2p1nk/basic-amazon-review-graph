# Amazon reviews:

sort the data based on if the reviews are positive or not-positive


```python
import pandas as pd

parsedData = pd.read_csv('./amazon.csv',header=0,sep=',')
```

```python
x = list(parsedData.iloc[:, 1])
y = list(parsedData.iloc[:,0])
def positive(x):
    if x == 1:
        return True
    return False


def nonPositive(x):
    if x == 0:
        return True
    return False

xp = list(filter(positive,x))

xn = list(filter(nonPositive,x))

xaxis = ['positive reviews','negative reviews']

yaxis = [len(xp),len(xn)]

print(xaxis)


```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize= (10,5))

plt.bar(xaxis,yaxis,color='blue',width=0.4)

plt.xlabel('reviews')
plt.ylabel('number of reviews')
plt.title('positive and non-positive amazon reviews')
```
